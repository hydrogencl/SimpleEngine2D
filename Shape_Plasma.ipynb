```python
for n in range(len(arr)):
    arr_x[n] = arr[n]/1.0
    arr_y[n] = math.e**(1.0/arr_x[n])
    #print(arr_x[n], arr_y[n])
PLT.plot(arr_x, arr_y)
```

```python
arr = list(range(1, round(math.pi*100)))
arr_x = [0.0 for i in range(len(arr))]
arr_y = [0.0 for i in range(len(arr))]
for n in range(len(arr)):
    arr_x[n] = arr[n]/100.0
    arr_y[n] = math.cos(arr_x[n])
    #print(arr_x[n], arr_y[n])
PLT.plot(arr_x, arr_y)
```

```python
arr = list(range(1, round(math.pi*100)))
arr_x = [0.0 for i in range(len(arr))]
arr_y = [0.0 for i in range(len(arr))]
for n in range(len(arr)):
    arr_x[n] = arr[n]/100.0
    arr_y[n] = 1 * math.e**(-1*arr_x[n])
    #print(arr_x[n], arr_y[n])
PLT.plot(arr_x, arr_y)
```

```python
arr = list(range(1, round(math.pi*100)))
arr_x = [0.0 for i in range(len(arr))]
arr_y1 = [0.0 for i in range(len(arr))]
arr_y2 = [0.0 for i in range(len(arr))]

a=[-2.0, 1.0 ]
b=[2.0, 1.0 ]
c=[3.0, 1.0 ]

for n in range(len(arr)):
    arr_x[n] = arr[n]/100.0
    arr_y1[n] = math.sin(a[0]*arr_x[n]) * b[0] * math.e**(-1*c[0]*arr_x[n])
    arr_y2[n] = math.sin(a[1]*arr_x[n]) * b[1] * math.e**(-1*c[1]*arr_x[n])
    #print(arr_x[n], arr_y[n])
fig = PLT.figure()
ax  = fig.add_subplot(111)
ax.plot(arr_x, arr_y1,"r")
ax.plot(arr_x, arr_y2,"r")
ax.set_ylim((-3,3))
#PLT.ylim = [0,10]
#PLT.plot(arr_x, arr_y)
```

```python
arr = list(range(1, round(math.pi*100)))
arr_x = [0.0 for i in range(len(arr))]
arr_y = [0.0 for i in range(len(arr))]
for n in range(len(arr)):
    arr_x[n] = arr[n]/100.0
    arr_y[n] = (math.sin(arr_x[n]) + math.sin(2*arr_x[n]))/2.0
    #print(arr_x[n], arr_y[n])


PLT.plot(arr_x, arr_y)
PLT.ylim = [0,10]
```

```python
arr = list(range(0, round(math.pi*3)))
arr_x = [0.0 for i in range(len(arr))]
arr_y = [0.0 for i in range(len(arr))]
for n in range(len(arr)):
    arr_x[n] = arr[n]/3.0
    arr_y[n] = math.sin(arr_x[n]) * math.e**(-1*arr_x[n])
    #print(arr_x[n], arr_y[n])
arr_y.append(0)
arr_x.append(arr_x[-1] + arr_x[-1] - arr_x[-2])

PLT.plot(arr_x, arr_y)
PLT.ylim = [0,10]
```
